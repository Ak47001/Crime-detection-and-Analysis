```python
#imp = SimpleImputer(strategy = "most_frequent")
imp = IterativeImputer(max_iter=10,  initial_strategy='mean', random_state=0)
df_filled = pd.DataFrame(imp.fit_transform(df), columns = df.columns) 
df_filled = df_filled.round()
#df_filled= df_filled.astype('category')
print(df_filled)
```

```python

# Feature matrix and output labels
X = df_filled.loc[:,['Location','Day','month','dayhalf','age','First M','Gender']].values
y = df_filled.loc[:, df_filled.columns == 'crimetype'].values.flatten()
print(X.shape)
print(y.shape)
```

```python
df_filled.columns
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import precision_score, recall_score, accuracy_score
```

```python
kf = KFold(n_splits = 5, shuffle = True, random_state = 1)

train_pred = []
train_actual = []

test_pred = []
test_actual = []
for train_idx, test_idx in kf.split(X, y):
    X_train, y_train = X[train_idx, :], y[train_idx]
    X_test, y_test = X[test_idx, :], y[test_idx]
   
    # Create decision tree classifer
    clf = DecisionTreeClassifier(class_weight="balanced",criterion = 'gini', random_state = 20)

    # Train decision tree classifier 
    model = clf.fit(X_train, y_train) 
    train_pred.extend(clf.predict(X_train))
    train_actual.extend(y_train)
    
    # Test decision tree classifier
    test_pred.extend(clf.predict(X_test))
    test_actual.extend(y_test)
```

```python
def evaluate(actual, predicted, prefix=""):
    precision = precision_score(actual, predicted, average='weighted')
    recall = recall_score(actual, predicted, average='weighted')
    accuracy = accuracy_score(actual, predicted)

    print("%s Precision: %.3f%%, Recall: %.3f%%, Accuracy: %.3f%%" % (prefix, precision * 100, recall * 100, accuracy * 100))
```

```python
evaluate(train_actual, train_pred, "Train")
evaluate(test_actual, test_pred, "Test")
```
